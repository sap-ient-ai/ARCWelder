```python
# 👁 All regions are on the LEFT half of the grid
# 🧲 As we know Y mirrors seed-symmetry, we know the regions for the 🔵 seed will be on the RIGHT half of the grid
# ❓ 🔮 🧪 Maybe each pixel on Y is determined by its nearest seed (i.e. it's that or 0)
# 📌 pair3 might be a good candidate to test this out as it has 3 seeds

# 👁 The first 3 regions follow a clear progression (an x-flipped L of size 1 3 5)
# 👁 The bottom of the regions is 0 2 4 6 8
# ❓ Can we figure out a most-simple generative process that creates these regions?

# 🧠 ✨ 💡 🔮 🧪 What if we keep generating the x-flipped L-shapes while confining our view to the left half of the grid?

# 🧲 💡 What if each seed generates a pattern thus, and each pixel looks to the pattern generated by its nearest seed

from Grid import Grid

def pattern(k):
    U = V = Grid(hw=k, fill=bool((k + 1) % 2))
    while V._.size > 0:
        V[:] = ~V
        V = V[:-1, :-1]
    return U

X, Y = train[0]
pattern(X.w)

# left_half = lambda U: Supergrid(U, ny=1, nx=2)[0, 0]
left_half = lambda U: U.split(nx=2)[0]
left_half(Y['⚪']) > left_half(pattern(X.w))
```

```python
def solve(X):
    X = X.copy
    Y = X * '⚫'

    P = pattern(X.h)

    for _ in range(4):
        X, Y = X['⤴'], Y['⤴']
        if (c := X[0, 0]) != '⚫':
            def win(p):
                k = X.h - 1
                others = [ p.dist2(o)
                    for o in [(0,k), (k,0), (k,k)]
                        if X[o] != '⚫'
                ]
                return P[p] and p.mag2 < min(others)
            Y[Y.locs | win] = c
    return Y

[Y @ solve(X)  for (X, Y) in train | test]
```
